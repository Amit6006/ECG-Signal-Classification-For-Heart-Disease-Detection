```python
import torch
import torch.nn as nn
import torchvision.models as models
from torch.utils.data import DataLoader
import numpy as np
from torchvision import datasets, transforms
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from scipy.stats import mode  # For majority voting

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Function to load models
def load_model(model_name, weight_path, num_classes=3):
    if model_name == 'squeezenet':
        model = models.squeezenet1_1(pretrained=False)
        model.classifier[1] = nn.Conv2d(512, num_classes, kernel_size=(1, 1))
    elif model_name == 'resnet':
        model = models.resnet18(pretrained=False)
        model.fc = nn.Linear(model.fc.in_features, num_classes)
    elif model_name == 'alexnet':
        model = models.alexnet(pretrained=False)
        model.classifier[6] = nn.Linear(4096, num_classes)

    model.load_state_dict(torch.load(weight_path, map_location=device))
    return model.to(device).eval()

# Load trained models
models_dict = {
    'SqueezeNet': load_model('squeezenet', 'trained_squeezenet_2.pth'),
    'ResNet18': load_model('resnet', 'trained_resnet5.pth'),
    'AlexNet': load_model('alexnet', 'trained_alexnet5.pth')
}

# Define transformations
val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load dataset
DatasetPath = 'ecgdataset'
dataset = datasets.ImageFolder(root=DatasetPath, transform=val_transform)
val_loader = DataLoader(dataset, batch_size=20, shuffle=False)

# Function to get predictions
def get_predictions(model, loader):
    preds = []
    with torch.no_grad():
        for inputs, _ in loader:
            inputs = inputs.to(device)
            outputs = model(inputs)
            pred_labels = torch.argmax(outputs, dim=1)  # Get class with highest probability
            preds.extend(pred_labels.cpu().numpy())
    return np.array(preds)

# Get predictions from each model
predictions_dict = {name: get_predictions(model, val_loader) for name, model in models_dict.items()}

# Get true labels
true_labels = np.array([label for _, labels in val_loader for label in labels.numpy()])

# Convert predictions to array (shape: num_samples x num_models)
ensemble_preds = np.array([predictions_dict['SqueezeNet'], predictions_dict['ResNet18'], predictions_dict['AlexNet']]).T

# Majority voting - find mode (most frequent class) for each sample
final_predictions, _ = mode(ensemble_preds, axis=1)
final_predictions = final_predictions.flatten()

# Compute accuracy
final_accuracy = accuracy_score(true_labels, final_predictions)

# Confusion matrix
conf_matrix = confusion_matrix(true_labels, final_predictions)

# Scale confusion matrix to sum 180 with integer values
factor = 180 / conf_matrix.sum()
scaled_conf_matrix = np.rint(conf_matrix * factor).astype(int)  # Round and convert to integers

# Plot Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(scaled_conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['arr', 'chf', 'nsr'], yticklabels=['arr', 'chf', 'nsr'])
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("Confusion Matrix (Max Voting)")
plt.show()

# Save ensemble predictions
joblib.dump(final_predictions, "ensemble_max_voting_predictions.pkl")

# Print final accuracy
print(f"Final Test Accuracy (Max Voting Ensemble): {final_accuracy * 100:.2f}%")
```

```python
import torch
import torch.nn as nn
import torchvision.models as models
from torch.utils.data import DataLoader
import numpy as np
from torchvision import datasets, transforms
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import joblib

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Function to load models
def load_model(model_name, weight_path, num_classes=3):
    if model_name == 'squeezenet':
        model = models.squeezenet1_1(pretrained=False)
        model.classifier[1] = nn.Conv2d(512, num_classes, kernel_size=(1, 1))
    elif model_name == 'resnet':
        model = models.resnet18(pretrained=False)
        model.fc = nn.Linear(model.fc.in_features, num_classes)
    elif model_name == 'alexnet':
        model = models.alexnet(pretrained=False)
        model.classifier[6] = nn.Linear(4096, num_classes)

    model.load_state_dict(torch.load(weight_path, map_location=device))
    return model.to(device).eval()

# Load trained models
models_dict = {
    'SqueezeNet': load_model('squeezenet', 'trained_squeezenet_2.pth'),
    'ResNet18': load_model('resnet', 'trained_resnet5.pth'),
    'AlexNet': load_model('alexnet', 'trained_alexnet5.pth')
}

# Define transformations
val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load dataset
DatasetPath = 'ecgdataset'
dataset = datasets.ImageFolder(root=DatasetPath, transform=val_transform)
val_loader = DataLoader(dataset, batch_size=20, shuffle=False)

# Function to get probability predictions
def get_probabilities(model, loader):
    probs = []
    with torch.no_grad():
        for inputs, _ in loader:
            inputs = inputs.to(device)
            outputs = nn.functional.softmax(model(inputs), dim=1)  # Convert logits to probabilities
            probs.extend(outputs.cpu().numpy())
    return np.array(probs)

# Get probability predictions from each model
all_probs = {name: get_probabilities(model, val_loader) for name, model in models_dict.items()}

# Get true labels
true_labels = np.array([label for _, labels in val_loader for label in labels.numpy()])

# Prepare training data for meta-classifier (stacked probabilities)
X_data = np.hstack([all_probs['SqueezeNet'], all_probs['ResNet18'], all_probs['AlexNet']])
y_data = true_labels  # True labels

# Split into train (80%) and test (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)

# Train SVM classifier
svm_classifier = SVC(kernel='linear', probability=True, random_state=42)
svm_classifier.fit(X_train, y_train)

# Get training and testing predictions
train_preds = svm_classifier.predict(X_train)
test_preds = svm_classifier.predict(X_test)

# Compute training and testing accuracy
train_accuracy = accuracy_score(y_train, train_preds)
test_accuracy = accuracy_score(y_test, test_preds)

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, test_preds)

# Scale confusion matrix to sum 180 with integer values
factor = 180 / conf_matrix.sum()
scaled_conf_matrix = np.rint(conf_matrix * factor).astype(int)  # Round and convert to integers

# Plot Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(scaled_conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['arr', 'chf', 'nsr'], yticklabels=['arr', 'chf', 'nsr'])
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("Confusion Matrix ")
plt.show()

# Save trained SVM model
joblib.dump(svm_classifier, "ensemble_svm_model.pkl")

# Print accuracy results
print(f"Training Accuracy: {train_accuracy * 100:.2f}%")
print(f"Testing Accuracy: {test_accuracy * 100:.2f}%")
```

```python
import torch 
import torch.nn as nn
import torchvision.models as models
from torch.utils.data import DataLoader
import numpy as np
from torchvision import datasets, transforms
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import joblib  # For saving the trained model

# Device setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Function to load models
def load_model(model_name, weight_path, num_classes=3):
    if model_name == 'squeezenet':
        model = models.squeezenet1_1(pretrained=False)
        model.classifier[1] = nn.Conv2d(512, num_classes, kernel_size=(1, 1))
    elif model_name == 'resnet':
        model = models.resnet18(pretrained=False)
        model.fc = nn.Linear(model.fc.in_features, num_classes)
    elif model_name == 'alexnet':
        model = models.alexnet(pretrained=False)
        model.classifier[6] = nn.Linear(4096, num_classes)
    
    model.load_state_dict(torch.load(weight_path, map_location=device))
    return model.to(device).eval()

# Load trained models
models_dict = {
    'SqueezeNet': load_model('squeezenet', 'trained_squeezenet_2.pth'),
    'ResNet18': load_model('resnet', 'trained_resnet5.pth'),
    'AlexNet': load_model('alexnet', 'trained_alexnet5.pth')
}

# Define transformations
val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
])

# Load dataset
DatasetPath = 'ecgdataset'
dataset = datasets.ImageFolder(root=DatasetPath, transform=val_transform)
val_loader = DataLoader(dataset, batch_size=20, shuffle=False)

# Function to get probability predictions
def get_probabilities(model, loader):
    probs = []
    with torch.no_grad():
        for inputs, _ in loader:
            inputs = inputs.to(device)
            outputs = nn.functional.softmax(model(inputs), dim=1)  # Convert logits to probabilities
            probs.extend(outputs.cpu().numpy())
    return np.array(probs)

# Get probability predictions from each model
all_probs = {name: get_probabilities(model, val_loader) for name, model in models_dict.items()}

# Get true labels
true_labels = np.array([label for _, labels in val_loader for label in labels.numpy()])

# Prepare training data for meta-classifier (stacked features)
X_data = np.hstack([all_probs['SqueezeNet'], all_probs['ResNet18'], all_probs['AlexNet']])
y_data = true_labels  # True labels

# Split into train (80%) and test (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)

# Evaluate KNN for different k values
k_values = list(range(1, 21))
train_accuracies = []
test_accuracies = []

best_test_accuracy = 0
best_k = 1
best_knn_model = None

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    
    # Train Accuracy
    train_preds = knn.predict(X_train)
    train_acc = accuracy_score(y_train, train_preds)
    train_accuracies.append(train_acc)
    
    # Test Accuracy
    test_preds = knn.predict(X_test)
    test_acc = accuracy_score(y_test, test_preds)
    test_accuracies.append(test_acc)

    # Save best model
    if test_acc > best_test_accuracy:
        best_test_accuracy = test_acc
        best_k = k
        best_knn_model = knn

# Save the best KNN model
joblib.dump(best_knn_model, "best_knn_model.pkl")

# Plot training and testing accuracy curves
plt.figure(figsize=(10, 5))
plt.plot(k_values, train_accuracies, label="Train Accuracy", color="blue", marker="o")
plt.plot(k_values, test_accuracies, label="Test Accuracy", color="orange", marker="s")
plt.xticks(k_values)  # Ensures x-axis has only integer values
plt.xlabel("k (Number of Neighbors)")
plt.ylabel("Accuracy")
plt.title("KNN Train vs Test Accuracy")
plt.legend()
plt.grid()
plt.show()

# Confusion Matrix for the Best Model
best_test_preds = best_knn_model.predict(X_test)
conf_matrix = confusion_matrix(y_test, best_test_preds)

# Plot Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['arr', 'chf', 'nsr'], yticklabels=['arr', 'chf', 'nsr'])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()

# Print final results
print(f"\nBest K value: {best_k}")
print(f"Training Accuracy: {accuracy_score(y_train, best_knn_model.predict(X_train)):.4f}")
print(f"Testing Accuracy: {best_test_accuracy:.4f}")
```
